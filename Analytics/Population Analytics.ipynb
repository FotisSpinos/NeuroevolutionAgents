```python
import os
```

```python
import sys
```

```python
from matplotlib import pyplot as plt
```

```python
POPULATION_FILE = "pop_analytics.txt"

#sources
#https://stackabuse.com/read-a-file-line-by-line-in-python/

# get the current directory
path = os.getcwd()
```

```python
try:
    analytics_file = open(POPULATION_FILE)
except:
    print("Population analytics file could not open")
    sys.exit()
    
print('file found: ' + POPULATION_FILE)
analytics_file.seek(0)
```

```python
# store fitness values from file
fitness_items = []
best_fitness_items = []
avairage_attacks = []
line = analytics_file.readline()

while line:
    contents = ""
    desc = ""
    
    if ":" in line:
            seperator_index = line.index(':')
            desc = line[0 : seperator_index]
            
            contents = line[seperator_index + 2 : len(line) - 1]
            
            if desc == "population fitness":
                fitness_items.append(float(contents))
            elif desc == "best generation fitness":
                best_fitness_items.append(float(contents))
            elif desc == 'avairage attacks' :
                avairage_attacks.append(float(contents))
    
    # read next line
    line = analytics_file.readline()

# close the file
analytics_file.close()
```

```python
# display graph using matplotlib

plt.plot(fitness_items)
plt.title('fitness graph')
plt.ylabel('fitness values')
plt.xlabel('generation number')
plt.show()
```

```python
# display best fitness graph

plt.plot(best_fitness_items)
plt.title('best fitness graph')
plt.xlabel('generation number')
plt.ylabel('best fitness values')
plt.show()
```

```python
# display the highest best_fitness_item

highest_fitness = best_fitness_items[0]

for fitness_item in best_fitness_items:
    if fitness_item > highest_fitness:
        highest_fitness= fitness_item
        
print('Highest agent fitness: ' + str(highest_fitness))
```

```python
# graph avairage attacks on each population

plt.plot(avairage_attacks)
plt.title('avaieage attacks graph')
plt.xlabel('generation number')
plt.ylabel('avairage attacks')
plt.show()
```
